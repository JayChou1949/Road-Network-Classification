## How to easily generate CRHDs?

This section showcases some very easy-to-use methods to generate Colored Road Hierarchy Diagrams (CRHDs) through the functions from our `crhd_generator`.

1. Use `PlotCRHD` function to plot a single CRHD with central coordinates and radius.  
**Parameters:**  
* *center_point*: Tuple of central coordinates -> (lng, lat).  
* *dist*: Radius in meter -> int.  
* *name*: Name of the plotted urban zone -> str.  
* *save_path*: The save path of the CRHD image. If None, the CRHD would be shown inline.  
* *dpi*: Image dpi.  
* *format*: Image format.  

return: If save_path is None, return the CRHD image. If save_path is not None, return None.  

```python
from crhd_generator import PlotCRHD
```

```python
# Plot the CRHD image inline

img = PlotCRHD(center_point=(41.88410159,-87.63208459), 
               dist=1000, 
               save_path=None
              )
```

```python
# Save the CRHD image to a given directory.

PlotCRHD(center_point=(41.88410159,-87.63208459), 
         dist=1000, 
         name='Chicago', 
         save_path='./data/'
        )
```

2. Use `PlotCity` function to generate and save CRHDs for a gridded city.

**Parameters:**  
* *dist*: radius in meter -> int.  
* *grid_path*: filepath of the grids Shapefile.  
* *save_path*: filepath to save the CRHD images.  
* *dpi*: image dpi.  
* *format*: image format.  

*return*: None.

```python
from crhd_generator import PlotCity
```

```python
# Generate and save CRHDs for a gridded city

grid_path = './data/Singapore/grids/grids.shp'
save_path = './data/Singapore/images/'

PlotCity(dist=1000, 
         grid_path=grid_path, 
         save_path=save_path
        )
```

After running this code, you should be able to see the CRHD images being generated one by one in your given save folder.
